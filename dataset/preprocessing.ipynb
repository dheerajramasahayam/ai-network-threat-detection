# UNSW-NB15 preprocessing

This notebook documents the preprocessing steps used by the research pipeline for the paper *AI-Based Network Threat Detection Using Deep Learning for Enterprise Infrastructure*.

```python
from pathlib import Path
import pandas as pd
from training.data_pipeline import engineer_unsw_features, load_unsw_nb15, summarize_unsw_nb15

REPO_DIR = Path.cwd().resolve()
if not (REPO_DIR / 'training').exists():
    REPO_DIR = REPO_DIR.parent
DATASET_DIR = REPO_DIR / 'dataset' / 'raw' / 'unsw_nb15'

summary = summarize_unsw_nb15(DATASET_DIR)
summary
```

```python
train_df, test_df = load_unsw_nb15(DATASET_DIR)
engineered_train = engineer_unsw_features(train_df)
engineered_train[['attack_cat', 'flow_duration', 'source_bytes', 'destination_bytes', 'avg_packet_size', 'bytes_per_second']].head()
```

```python
class_distribution = (
    pd.concat([train_df, test_df], ignore_index=True)['attack_cat']
    .fillna('Normal')
    .value_counts()
    .rename_axis('attack_cat')
    .reset_index(name='rows')
)
class_distribution
```
