# Feature engineering

This notebook explains how the paper converts raw UNSW-NB15 records into flow features that support Random Forest, LSTM, and Transformer models.

```python
from pathlib import Path
from training.data_pipeline import engineer_unsw_features, load_unsw_nb15

REPO_DIR = Path.cwd().resolve()
if not (REPO_DIR / 'training').exists():
    REPO_DIR = REPO_DIR.parent
DATASET_DIR = REPO_DIR / 'dataset' / 'raw' / 'unsw_nb15'

train_df, _ = load_unsw_nb15(DATASET_DIR)
features = engineer_unsw_features(train_df)
```

```python
feature_columns = [
    'flow_duration',
    'source_bytes',
    'destination_bytes',
    'total_bytes',
    'avg_packet_size',
    'bytes_per_second',
    'packets_per_second',
    'tcp_round_trip_time',
    'service_connection_count',
    'destination_connection_count',
]
features[feature_columns + ['attack_cat']].head()
```

```python
feature_summary = features[feature_columns].describe().T[['mean', 'std', 'min', 'max']]
feature_summary
```

The engineered feature set mixes packet volume, timing, transport state, and connection-behavior signals. This is the same representation used by the experiment runner in `training/run_experiments.py`.
